```python
import pandas as pd
import numpy as np
import re
```

```python
sales_test = pd.read_excel("../../sales_test.xlsx",header =1)
sales_test
```

```python
sales_test = sales_test[['상품명', '상품군']].drop_duplicates(subset=['상품명'])
sales_test
```

```python
sales_test = sales_test[sales_test['상품군'] != '무형']
sales_test
```

```python
sales_test.reset_index(drop=True, inplace=True)
sales_test
```

```python
sales_test['cat_1'] = None
sales_test['cat_2'] = None
sales_test['cat_3'] = None
sales_test
```

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
```

```python
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
driver = webdriver.Chrome(chrome_options=chrome_options, executable_path='/Users/nari/Downloads/chromedriver')
```

```python
def set_categories(df):

    for index, row in df.iterrows():

        name = re.sub(r'\([^)]*\)', '', row['상품명'])
        url = "https://www.coupang.com/np/search?component=&q={}&channel=user".format(name)
        driver.get(url)
        wait = WebDriverWait(driver, 10)

        try:
            if row['상품군'] == '가전':
                element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.search-product-list li:nth-child(2) a.search-product-link')))

            else:
                element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.search-product-list li.search-product a.search-product-link')))

            select_product_url = element.get_attribute('href')
            driver.get(select_product_url)

            elements = []

            try:
                cat1 = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#breadcrumb li:nth-child(2)')))
                cat2 = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#breadcrumb li:nth-child(3)')))
                cat3 = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#breadcrumb li:nth-child(4)')))

                df.at[index,'cat_1'] = cat1.text.split()[0]
                df.at[index,'cat_2'] = cat2.text.split()[0]
                df.at[index,'cat_3'] = cat3.text.split()[0]
#                 print('categories: {} > {} > {}'.format(cat1.text.split()[0], cat2.text.split()[0], cat3.text.split()[0]))


            except:
                # Not all categories found
                print('exception: 카테고리 없음: {}'.format(name))
                pass
                            
        except:
            # search not found
            print('exception: 상품검색 결과/추천상품 없음: {}'.format(name))
            pass
```

```python
set_categories(sales_test)
```
