```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
import datetime
import numpy as np
from sklearn.model_selection import train_test_split
import lightgbm as lgb 
import time 
from xgboost import XGBRegressor
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
```

```python
# 사용된 lgbm 모델 _ 하이퍼파라미터 튜닝은 아직 하지 못함
# 총취급액이 아니라 판매건수를 기준으로 학습 진행.
# 판매건수를 예측한 후 일정기준(magic_number) 을 넘지 못하면, 판매량이 0 인 것으로 처리

def lgbm_ca(table, magic_number):
    feature_columns = list(table.columns.difference(['count','취급액']))
    X = table[feature_columns].values
    y = table["count"].values

    train_X,test_X,train_y,test_y = train_test_split(X,y,test_size = 0.1, random_state = 35)

    print(train_X.shape,train_y.shape,test_X.shape,test_y.shape)
    
    lgbm_model = lgb.LGBMRegressor(n_estimators = 500, 
                                     learing_rate = 0.01, 
                                     max_depth = -1,
                                     boosting = 'gbdt',
                                    feature_fraction = 0.9,
                                     bagging_fraction = 0.7, 
                                     bagging_freq = 1,
                                    num_iteration = 10000, 
                                     objective = 'mape', 
                                     lambda_l2 = 0.1,
                                     max_bin =255,
                                    num_leaves = 1024, 
                                     min_date_in_leaf = 0)
    evals =[(test_X,test_y)]
    lgbm_model.fit(train_X,train_y,early_stopping_rounds =300, eval_metric = "mape", eval_set = evals, verbose =True)
    
    preds = lgbm_model.predict(test_X)
    prediction = lgbm_model.predict(X)
    
    preds = np.where(preds > magic_number , preds, 0)
    test_y_mother = np.where(test_y == 0, 1, test_y)

    return np.mean(np.abs((test_y - preds)/test_y_mother)), prediction
```
